```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("C:\\Users\\dharm\\Desktop\\DailyDelhiClimateTest.csv") # Uploading data..

data['date'] = pd.to_datetime(data['date']) 

data['month'] = data['date'].dt.month 
```

```python
data # print the Data
```

```python
data.describe() # calculate mean, median, max, min, count,standard deviation...
```

# Line plot

```python

plt.figure(figsize=(10, 6)) # figure size...

plt.plot(data['date'], data['meantemp'], color='orange', linewidth=2, label="Mean Temperature", marker='o', markersize=4) # plot the graph...

plt.title('Mean Temperature Over Time', fontsize=14, fontweight="bold", color='darkblue') # Add Title...

plt.xlabel('Date', fontsize=12, fontweight="bold", color='darkgreen') #Add x- label...

plt.ylabel('Mean Temperature (Â°C)', fontsize=12, fontweight="bold", color='darkgreen') #Add y - label...

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7) # Add Grid...

plt.legend(loc='upper left', fontsize=12, facecolor="white", edgecolor="black") # Add legend...

plt.xticks(rotation=45, fontweight="bold", fontsize=10, color='black') # Add X-Ticks....

plt.yticks(fontweight="bold", fontsize=10, color='black')# Add Y-ticks....

plt.gca().set_facecolor('lightyellow') 

average_temp = data['meantemp'].mean() 

plt.axhline(y=average_temp, color='black', linestyle='--', linewidth=1, label='Average Temperature') #

plt.legend(loc='upper left', fontsize=12, facecolor="white", edgecolor="black") # Add Legend


plt.tight_layout()


plt.savefig('mean_temperature_over_time.png')


plt.show() #Displays the Graph...
```

#  Bar plot

```python
avg_humidity = data.groupby('month')['humidity'].mean()

plt.figure(figsize=(10, 6))

bars = plt.bar(avg_humidity.index, avg_humidity.values, color='grey', edgecolor="black") # plot a graph

plt.title('Average Humidity by Month', fontsize=14, fontweight="bold", color='darkblue') # Add Title

plt.xlabel('Month', fontsize=12, fontweight="bold", color='darkgreen') # Add X- label

plt.ylabel('Average Humidity (%)', fontsize=12, fontweight="bold", color='darkgreen') #Add Y-label

plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45,fontweight = "bold")

plt.yticks(fontweight = "bold")
plt.grid(axis="both", color='gray', linestyle='--', linewidth=0.5, alpha=0.7) # Add Grid

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, round(yval, 1), ha='center', va='bottom', fontsize=10, fontweight='bold', color='orange') # Add Text and values 

overall_avg_humidity = avg_humidity.mean()

plt.axhline(y=overall_avg_humidity, color='red', linestyle='--', linewidth=1, label='Overall Average Humidity') #

plt.legend(loc='upper right', fontsize=12, facecolor="white", edgecolor="black") # Add legend

plt.tight_layout()

plt.savefig('average_humidity_by_month.png')

plt.show() # Display the Graph
```

# Scatter plot

```python

plt.figure(figsize=(10, 6)) # Figure size...

sns.scatterplot(x='meantemp', y='wind_speed', data=data, color='red', label="Wind Speed (m/s) vs Mean Temperature (Â°C)", s=30, alpha=0.7, edgecolor='black') # plot the Graph..

plt.title('Wind Speed vs Mean Temperature', fontsize=14, fontweight="bold", color='darkblue') # Add the Title.

plt.xlabel('Mean Temperature (Â°C)', fontsize=12, fontweight="bold", color='darkgreen') # Add X-label with size and color..

plt.ylabel('Wind Speed (m/s)', fontsize=12, fontweight="bold", color='darkgreen') # Add Y - label with size and color..

plt.xticks(fontweight = "bold")

plt.yticks (fontweight = "bold")

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7) # Add Grid with style..

plt.legend(loc='upper left', fontsize=10, facecolor="white", edgecolor="black") # Add Legend with font size and color..

plt.tight_layout() # Figure Adjustment...

plt.savefig('wind_speed_vs_mean_temperature.png') #Save the Figure...

plt.show() # Displays the Graph..
```

# Box Plot

```python
plt.figure(figsize=(10, 6)) # Figure Size...

sns.boxplot(y='meantemp', data=data, color='lightpink', fliersize=5) # plot the box plot

plt.title('Distribution of Mean Temperature', fontsize=14, fontweight="bold", color='darkblue') # Add Title with size and color...

plt.ylabel('Mean Temperature (Â°C)', fontsize=12, fontweight="bold", color='darkgreen') # Add Y- label with size and color...

plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5, alpha=0.7) # Add Grid lines with color..

plt.yticks(fontweight = "bold")

median_temp = data['meantemp'].median() # Calculate Median...

plt.text(0, median_temp, f'Median: {median_temp:.2f} Â°C', horizontalalignment='center', fontsize=10, color='blue', fontweight='bold') # Add Text...

plt.tight_layout() # Adjust the figure..

plt.savefig('mean_temperature_distribution.png') # Save the Figure...

plt.show() # Display the Graph...
```
